```python
import torch
from torch.utils.data import Dataset,DataLoader
from torchvision import transforms,models
from json import load
import cv2
import numpy as np
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
class keypoints_dataset(Dataset):
    def __init__(self,img_dir,data_file):
        self.img_dir = img_dir
        with open(data_file,'r') as f:
            self.data = load(f)
        self.transform = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((224,224)),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ])


    def __len__(self):
        return len(self.data)

    def __getitem__(self,index):
        item = self.data[index]
        image = cv2.imread(f"{self.img_dir}/{item['id']}.png")
        h,w = image.shape[:2]

        image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
        image = self.transform(image)
        keypoints = np.array(item['kps']).flatten()
        keypoints = keypoints.astype(np.float32)

        keypoints[::2] *= 224.0/w
        keypoints[1::2] *= 224.0/h
        return image,keypoints
```

```python
train_dataset = keypoints_dataset('/content/data/images', '/content/data/data_train.json')
valid_dataset = keypoints_dataset('/content/data/images', '/content/data/data_val.json')

train_loader = DataLoader(train_dataset,batch_size=8,shuffle=True)
valid_loader = DataLoader(valid_dataset,batch_size=8,shuffle=True)
```

```python
model = models.resnet50(pretrained=True)
model.fc = torch.nn.Linear(model.fc.in_features,14*2)
```

```python
model_resnet_custom = model.to(device)
```

```python
criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(),lr=1e-4)
```

```python
epochs = 20

for epoch in range(epochs):
    for i,(image,keypoints) in enumerate(train_loader):
        image = image.to(device)
        keypoints = keypoints.to(device)
        optimizer.zero_grad()
        output = model(image)
        loss = criterion(output,keypoints)
        loss.backward()
        optimizer.step()

        if i%10 == 0:
            print(f"Epoch: {epoch}, step: {i}, loss: {loss.item()}")
```
